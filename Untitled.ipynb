```python
import scipy.integrate as inte
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Functions with Translocation Strategy

# Donor Population equation

# inputs
# u = recruitment rate aka birth rate/increasing,
# y1 = population of donor population at t 
# k = carrying capacity 
# h = rate rhinos are being translocated

def donor_and_pred(t,x,u1,u2,u3,k1,k2,k3,h1,h2,p,p1,p2,p3,r1,r2,r3,q1,q2,q3):
  y1 = x[0]
  y2 = x[1]
  y3 = x[2]
  d1 = x[3]
  d2 = x[4]
  d3 = x[5]

  dy1dt = u1[int(t)]*y1*(1-(y1/k1[int(t)]))-h1*y1 - p1[int(t)]*d1*y1
  dy2dt = u2[int(t)]*y2*(1-(y2/k2[int(t)])) + p*h1*y1 - h2*y2 - p2[int(t)]*d2*y2
  dy3dt = u3[int(t)]*y3*(1-(y3/k3)) + (1-p)*h1*y1 + h2*y2 - p3[int(t)]*d3*y3

  dd1dt = r1*y1*d1 - q1*d1
  dd2dt = r2*y2*d2 - q2*d2
  dd3dt = r3*y3*d3 - q3*d3

  return np.array([dy1dt, dy2dt, dy3dt, dd1dt, dd2dt, dd3dt])
```

```python
from scipy.integrate._ivp.rk import RK23
p = 0.7
#k1 = 200+20*np.sin((np.pi*t)/5)
#k2 = 50+5*np.sin((np.pi*5)/5)
k3 = 60
h1 = 0.01
h2 = 0.01
r1 = 0.05 # https://rhinos.org/about-rhinos/state-of-the-rhino/#:~:text=Officials%20are%20forecasting%20a%20continued,threat%20to%20wildlife%20in%20Nepal.
r2 = 0.05
r3 = 0.05
q1 = 0.3356 #rate of unemployment in South Africa 2021 https://www.statista.com/statistics/370516/unemployment-rate-in-south-africa/
q2 = 0.3356
q3 = 0.3356 

init_cond = [180,2,2,19,19,19]
trange=np.array([0, 15])
tlist=np.linspace(0, 15, 200)

k1 = 200+20*np.sin((np.pi*tlist)/5)
k2 = 50+5*np.sin((np.pi*tlist)/10)
p1 = np.concatenate((tlist[0:100], tlist[100:0:-1])) # https://www.savetherhino.org/rhino-info/poaching-stats/
p1 = 0.08*np.heaviside(p1-5,0)+0.02
p2 = p1
p3 = p1
u1 = 0.05*np.sin((tlist/1.6)-1.64)+0.06
u2 = 0.05*np.sin((tlist/1.6)-1.64)+0.06
u3 = 0.05*np.sin((tlist/1.6)-1.64)+0.06

sol = inte.solve_ivp(donor_and_pred, trange, init_cond, t_eval=tlist, args=(u1,u2,u3,k1,k2,k3,h1,h2,p,p1,p2,p3,r1,r2,r3,q1,q2,q3))
```
