```python
import matplotlib.pylab as plt   # That gives plotting, and the next line makes plots appear inline in notebook
%matplotlib inline  
import numpy as np  # That gives numerical arrays and tools for manipulating them
import scipy.optimize as opt
import scipy.linalg as la
import pandas as pd
from random import seed
from random import random
Ai = np.array([[0.98, 0.1, 0],[0.02, 0.7, 0.05], [0, 0.2, 0.95]])
print(Ai)
Ao = np.array([[0.90, 0.1, 0],
             [0.10, 0.6, 0.1], 
             [0, 0.3, 0.9]])
print(Ao)
```

```python
Ni = 100
No = 50
```

```python
state=np.array([[1.0, 0.0, 0.0]])
stateHist=state
dfStateHist=pd.DataFrame(state)
distr_hist = [[0,0,0]]
for x in range(Ni):
  state=np.dot(state,Ai)
  #print(state)
  stateHist=np.append(stateHist,state,axis=0)
  dfDistrHist = pd.DataFrame(stateHist)
  dfDistrHist.plot()
plt.show()
```
